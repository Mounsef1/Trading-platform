```python
from bs4 import BeautifulSoup
import requests

def explore_cnn_structure():
    url = "https://www.cnn.com/"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find all anchor tags as a broad search to examine titles and links
    for link in soup.find_all("a"):
        text = link.get_text().strip()
        href = link.get("href")
        if text and href:  # Only print if both title and link are available
            print(f"Text: {text}, Link: {'https://www.cnn.com' + href if href.startswith('/') else href}")

def explore_bbc_structure():
    url = "https://www.bbc.com/"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find all anchor tags to examine titles and links
    for link in soup.find_all("a"):
        text = link.get_text().strip()
        href = link.get("href")
        if text and href:
            print(f"Text: {text}, Link: {'https://www.bbc.com' + href if href.startswith('/') else href}")

# Run tests
explore_cnn_structure()
explore_bbc_structure()
```

```python
from bs4 import BeautifulSoup
import requests

def get_distinct_cnn_articles():
    url = "https://www.cnn.com/"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Keywords to detect newsworthy articles, and terms to exclude general links
    article_keywords = ["analysis", "live updates", "why", "reveals", "finds", "report", "says", "shows", "intentionally", "risk"]
    ignored_keywords = ["newsletter", "video", "live tv", "settings", "topics you follow", "sign out", "terms", "privacy", "about"]
    articles = {}

    for link in soup.find_all("a"):
        text = link.get_text(strip=True)
        href = link.get("href")
        
        if text and href and len(text) > 10:
            # Ensure it contains specific keywords and exclude general or non-article links
            if any(word in text.lower() for word in article_keywords) and not any(kw in href.lower() for kw in ignored_keywords):
                full_link = f"https://www.cnn.com{href}" if href.startswith("/") else href
                if text not in articles:  # Avoid duplicates
                    articles[text] = full_link

    for title, url in articles.items():
        print(f"Title: {title}, Link: {url}")

# Run optimized script
get_distinct_cnn_articles()
```

```python
import pandas as pd
from urllib.parse import urlparse
from crawlee.playwright_crawler import PlaywrightCrawler, PlaywrightCrawlingContext
from bs4 import BeautifulSoup

class Crawler:
    def __init__(self, seed_url):
        self.seed_url = seed_url
        self.allowed_domain = urlparse(seed_url).netloc.lower()
        self.data_frame = pd.DataFrame(columns=["url", "title", "text"])  # Initialize empty DataFrame

    def crawl(self) -> None:
        crawler = PlaywrightCrawler()

        @crawler.router.default_handler
        def request_handler(context: PlaywrightCrawlingContext) -> None:
            current_domain = urlparse(context.request.url).netloc.lower()

            # Skip processing if the current page is from an external domain
            if current_domain != self.allowed_domain and not current_domain.endswith(f".{self.allowed_domain}"):
                print(f"Skipping: {context.request.url} (outside allowed domain)")
                return

            # Extract the HTML content of the page
            page = context.page
            title = page.title()
            print("Page title:", title)
            html_content = page.content()

            # Parse the HTML content with BeautifulSoup
            soup = BeautifulSoup(html_content, 'html.parser')
            print("HTML content:", html_content[:500])  # Print first 500 characters of HTML


            # Remove script and style elements
            for script_or_style in soup(["script", "style"]):
                script_or_style.extract()  

            # Extract and clean visible text
            visible_text = ' '.join(soup.get_text(separator=' ', strip=True).split())
            print("Visible text:", visible_text[:500])

            # Collect data
            data = {
                "url": context.request.url,
                "title": page.title(),
                "text": visible_text.replace(",", "")
            }
            print("Data to append:", data)
            
            # Append the data to the DataFrame
            self.data_frame = pd.concat([self.data_frame, pd.DataFrame([data])], ignore_index=True)
            self.data_frame.to_csv("temp_output.csv", index=False)

            # Log progress
            print(f"Processing: {context.request.url}")

            # Filter function to limit links to the allowed domain
            def filter_func(link):
                link_domain = urlparse(link.url).netloc.lower()
                if link_domain.startswith("www."):
                    link_domain = link_domain[4:]
                return link_domain == self.allowed_domain or link_domain.endswith(f".{self.allowed_domain}")

            # Enqueue links within the allowed domain
            context.enqueue_links(filter_func=filter_func)

        # Start crawling from the initial seed URL
        crawler.run([self.seed_url])

        # Save the data to a CSV file
        logfile = f"{urlparse(self.seed_url).netloc.lower()}.csv"
        self.data_frame.to_csv(logfile, index=False)
        print(f"Data saved to {logfile}")

# Usage example
if __name__ == "__main__":
    seed_url = "https://cnn.com/"
    crawler_instance = Crawler(seed_url)
    crawler_instance.crawl()
```
